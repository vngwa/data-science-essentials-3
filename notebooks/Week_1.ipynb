# Create a calculated column in the income_county_agg DataFrame which gives the approximate average household income for each county. Calculate this as 1000*total_inc_amt / return_count.

#assigning a new column avg_household_income to the income_county_agg DataFrame

```python
income_county_agg_add_col= income_county_agg[['county','total_inc_amt','return_count']]
income_county_agg_add_col=income_county_agg_add_col.assign(avg_household_income=(income_county_agg['total_inc_amt']/income_county_agg['return_count'].sum())*1000)
income_county_agg_add_col
```

#Create a histogram showing the distribution of average incomes across all counties in Tennessee. Be sure to remove the Tennessee row prior to creating this histogram

```python
income_county_agg_add_col.hist(column='avg_household_income')
```
